# Coba Hyperparameter

```python
train = pd.read_csv('data_train_clean.csv')
test = pd.read_csv('data_test_clean.csv')
```

```python
train.shape
```

```python
test.shape
```

```python
train
```

```python
X_train = train.drop(columns=['prognosa'])
y_train = train["prognosa"]
X_test = test.drop(columns=['prognosa'])
y_test = test["prognosa"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
```

```python
rf = RandomForestClassifier()
```

```python
# Daftar hyperparameter yang akan diuji
param_grid = {
    'n_estimators': [100, 200, 300],  # Jumlah pohon dalam random forest
    'max_depth': [None, 5, 10],  # Kedalaman maksimum setiap pohon
    'min_samples_split': [2, 5, 10],  # Jumlah sampel minimum yang diperlukan untuk membagi node internal
    'min_samples_leaf': [1, 2, 4]  # Jumlah sampel minimum yang diperlukan untuk menjadi daun pada pohon
}
```

```python
# Membuat objek GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
```

```python
# Melakukan pelatihan pada data train
grid_search.fit(X_train, y_train)
```

```python
# Mencetak hyperparameter terbaik yang ditemukan
print("Hyperparameter terbaik: ", grid_search.best_params_)
```

```python
# Mencetak akurasi model terbaik pada data uji
y_pred = grid_search.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model terbaik: {:.2f}%".format(accuracy * 100))
```

```python
print(classification_report(y_test, y_pred))
```
